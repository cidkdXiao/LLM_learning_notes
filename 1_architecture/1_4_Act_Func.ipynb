## Sigmoid 函数

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 sigmoid 函数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 生成数据点
x = np.linspace(-10, 10, 500)  # 从 -10 到 10 之间取 500 个点
y = sigmoid(x)

# 绘制图像
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="sigmoid(x)", color="blue")
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# 保存图像到指定路径
plt.savefig("sigmoid.png", dpi=300, bbox_inches="tight")  # 这里可以改成你想要的路径，比如 "D:/plots/sigmoid.png"

plt.show()
```

## Tanh

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 tanh 函数
def tanh(x):
    return np.tanh(x)

# 生成数据点
x = np.linspace(-5, 5, 500)  # 从 -5 到 5 之间取 500 个点
y = tanh(x)

# 绘制图像
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="tanh(x)", color="red")
plt.title("Tanh Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.savefig("tanh.png", dpi=300, bbox_inches="tight")

# 显示图像
plt.show()
```

## ReLU

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 ReLU 函数
def relu(x):
    return np.maximum(0, x)

# 生成数据点
x = np.linspace(-5, 5, 500)  # 从 -5 到 5 之间取 500 个点
y = relu(x)

# 绘制图像
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="ReLU(x)", color="green")
plt.title("ReLU Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.savefig("relu.png", dpi=300, bbox_inches="tight")

# 显示图像
plt.show()
```

## Leaky ReLU

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 Leaky ReLU 函数
def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)

# 生成数据点
x = np.linspace(-5, 5, 500)  # 从 -5 到 5 之间取 500 个点
y = leaky_relu(x, alpha=0.1)  # 这里设置 alpha=0.1，你可以改成 0.01 或其他值

# 绘制图像
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Leaky ReLU(x)", color="purple")
plt.title("Leaky ReLU Function")
plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.savefig("leaky_relu.png", dpi=300, bbox_inches="tight")

# 显示图像
plt.show()
```

## ELU

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 ELU 函数
def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

# 生成数据点
x = np.linspace(-5, 5, 500)  # 从 -5 到 5 之间取 500 个点
y = elu(x, alpha=1.0)        # alpha 可调，常用值是 1.0

# 绘制图像
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="ELU(x)", color="orange")
plt.title("ELU Function (alpha=1.0)")
plt.xlabel("x")
plt.ylabel("ELU(x)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.savefig("elu.png", dpi=300, bbox_inches="tight")

# 显示图像
plt.show()
```

## Swish

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid 函数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Swish 函数
def swish(x, beta):
    return x * sigmoid(beta * x)

# 生成数据点
x = np.linspace(-5, 5, 500)

# 三种情况
y_beta0 = swish(x, beta=0)          # 相当于 0.5 * x
y_beta1 = swish(x, beta=1)          # 标准 Swish
y_betainf = np.maximum(0, x)        # beta→∞ 时趋近 ReLU

# 绘制图像
plt.figure(figsize=(8, 5))
plt.plot(x, y_beta0, label="Swish (β=0)", linestyle="--", color="blue")
plt.plot(x, y_beta1, label="Swish (β=1)", linestyle="-", color="green")
plt.plot(x, y_betainf, label="Swish (β→∞ ≈ ReLU)", linestyle="-.", color="red")

plt.title("Swish Function with Different β")
plt.xlabel("x")
plt.ylabel("Swish(x)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

plt.savefig("swish.png", dpi=300, bbox_inches="tight")
plt.show()
```
